```python
import numpy as np
import glob
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from scipy.misc import imread, imsave, imresize
```

```python
def load_images_with_label(imagePath, extension, imageArray=None,labelArray=None, label='nothing'):
    img_rows = 100
    img_cols = 100
    filelist = glob.glob(imagePath + '/*.' + extension)
    x = np.zeros((len(filelist), img_rows, img_cols))
    y = [None]*len(filelist)
    for i, filepath in enumerate(filelist):
        image = imread(filepath, True) 
        imageResized = imresize(image, (img_rows, img_cols))
        x[i] = imageResized
        y[i] = label
    
    if imageArray is not None:
        x = np.concatenate((imageArray, x), axis=0)
        labelArray.append(y)
        y = labelArray
    return x, y
```

```python
x, y = load_images_with_label(label='word', imagePath='words', extension='png')
print x.shape
```
